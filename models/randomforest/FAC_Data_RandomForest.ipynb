```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

data = pd.read_csv("FAC_ML_Train.csv")

features = ["going_concern_num", "material_weakness_num", "material_weakness_gen_num", "repeat_finding_num", "questioned_costs_num", "significant_deficiency_num", "modified_opinion_num", "other_findings_num", "other_matters_num", "noncompliance_num", "prior_findings_num", "low_risk_num", "has_cap_num"]

X = data[features].apply(pd.to_numeric, errors='coerce').fillna(0)
y = pd.to_numeric(data['is_high_risk'], errors='coerce').fillna(0)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)

importance = pd.DataFrame({'Variable': features, 'Importance': rf.feature_importances_}).sort_values('Importance', ascending=False)

print(importance)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(rf, X, y, cv=5, scoring='accuracy')
print(f"Accuracy: {scores.mean():.2%}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Get predictions
y_pred = rf.predict(X)

# Confusion Matrix
cm = confusion_matrix(y, y_pred)
print("Confusion Matrix:")
print(cm)

# Precision, Recall, F1
print("\nClassification Report:")
print(classification_report(y, y_pred, target_names=['Low/Moderate Risk', 'High Risk']))
```
